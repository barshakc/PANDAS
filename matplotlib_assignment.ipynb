# **LEARNING MATPLOTLIB**

**Q.1.LINE PLOTTING**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
a=[2,4,6,8,10,11,11.5,11.7]
b=[1,1.5,2,2.5,3,3.5,4,4.5]
pyplot.grid()
pyplot.title('VISUALIZING DEPENDENCY OF X-LABEL AND Y-LABEL',fontsize=25)
pyplot.xlabel('x_label',fontsize=12)
pyplot.ylabel('y_label',fontsize=12)
pyplot.legend(['x_label','y_label'])
pyplot.plot(a,b,color='red')
pyplot.show()
```

**Q.3.BAR GRAPH**

```python
from matplotlib import pyplot as plt
x=['2016','2017','2018','2019','202']
y=[1252,1632,1692,1856,865]
plt.title('DUMMY MOVIES INFO',fontsize=25)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Number of Movies Released',fontsize=12)
plt.bar(x,y,color='#4CAF50')
plt.show()
```

**Q.4.PIE PLOTTING**

```python
import matplotlib.pyplot as plt
labels='Potato','Steak','Bread','Milk'
sizes=[10,25,45,20]
explode=(0.01,0.01,0.01,0.1)
plt.pie(sizes,labels=labels,startangle=90,explode=explode,shadow=True)
plt.legend()
plt.show()
```

**Q.7.USING SUBPLOT FUNCTION**



*   **HORIZONTALLY**



```python
import numpy as np 
import matplotlib.pyplot as plt  
x = np.arange(0, 3 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x) 
print('HORIZONTAL STACK')
fig,ax=plt.subplots(2,1)
ax[0].plot(x,y_sin,color='black')
ax[1].plot(x,y_cos,color='black')
plt.show()
```



* VERTICALLY




```python
import numpy as np 
import matplotlib.pyplot as plt  
x = np.arange(0, 3 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x) 
print('VERTICAL STACK')
fig,ax=plt.subplots(1,2)
ax[0].plot(x,y_sin,color='c')
ax[1].plot(x,y_cos,color='c')
plt.show()
```
