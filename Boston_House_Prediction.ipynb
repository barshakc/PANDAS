```python
#creating box plots for all attributes
fig,ax=plt.subplots(ncols=7,nrows=2,figsize=(20,10))
index=0
ax=ax.flatten()

for col,value in hd.items():
  sns.boxplot(y=col,data=hd,ax=ax[index])
  index+=1
  #There is some overlapping so we will do hyperparameter tuning
  plt.tight_layout(pad=0.5,w_pad=0.7,h_pad=5.0)
```

```python
#creating distplot
fig,ax=plt.subplots(ncols=7,nrows=2,figsize=(20,10))
index=0
ax=ax.flatten()

for col,value in hd.items():
  sns.distplot(value,ax=ax[index])
  index+=1
  plt.tight_layout(pad=0.5,w_pad=0.7,h_pad=5.0)
```

```python
#using minmax normalization to those columns which have larger values
cols=['CRIM','ZN','TAX','B']
for col in cols:
  #finding minimum and maximum of that column
    minimum=min(hd[col])
    maximum=max(hd[col])
    hd[col]=(hd[col]-minimum)/(maximum-minimum)
#Again creating distplot
fig,ax=plt.subplots(ncols=7,nrows=2,figsize=(20,10))
index=0
ax=ax.flatten()

for col,value in hd.items():
  sns.distplot(value,ax=ax[index])
  index+=1
  plt.tight_layout(pad=0.5,w_pad=0.7,h_pad=5.0)


```



> **Now, we got the values within the same range**


**STANDARDIZATION**

```python
from sklearn import preprocessing
scalar=preprocessing.StandardScaler()

scaled_cols=scalar.fit_transform(hd[cols])
scaled_cols=pd.DataFrame(scaled_cols,columns=cols)
scaled_cols.head()
```

```python
for col in cols:
  hd[col]=scaled_cols[col]
```

**SPLITTING THE DATA AND TARGET**

```python
X=hd.drop(['PRICE'],axis=1)
Y=hd['PRICE']
print(X)
print(Y)
```

**TRAIN-TEST SPLIT**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

**MODEL TRAINING**



1.   LINEAR REGRESSION



```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
LinearRegression(copy_X=True, fit_intercept=True ,normalize=False)
y_pred=model.predict(X_train)
print("Training accuracy:",model.score(X_train,Y_train)*100)
```

```python
print("Training accuracy:",model.score(X_test,Y_test)*100)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("Model accuracy:",r2_score(Y,model.predict(X))*100)
```

**MODEL VISUALIZATION**

```python
plt.scatter(Y_train,y_pred)
plt.xlabel=("PRICES")
plt.ylabel=("PREDICTED PRICES")
plt.title("PRICES VS PREDICTED PRICES")
plt.show()
```

2.   RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor()
reg.fit(X_train,Y_train)
y_pred=reg.predict(X_train)
print("Training accuracy:",reg.score(X_train,Y_train)*100)
```

```python
print("Testing accuracy:",reg.score(X_test,Y_test)*100)
```

**MODEL VISUALIZATION**

```python
plt.scatter(Y_train,y_pred)
plt.xlabel=("PRICES")
plt.ylabel=("PREDICTED PRICES")
plt.title("PRICES VS PREDICTED PRICES")
plt.show()
```
